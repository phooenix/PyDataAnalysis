```python
class Chocolate:
  def __init__(self, price: float, additives: str= ''):

    if price > 0:
      self.price = price
    else:
      raise ValueError('Цена должна быть больше 0')

    if type(additives) is str:
      self.additives = additives
    else:
      self.additives = ''

  def show_my_chocolate(self):
    if self.additives == '':
      print('Обычный шоколадный батончик')
    else:
      print('Шоколадный батончик, добавка: ' + self.additives)

  def __str__(self):
    return f'Шоколадный батончик, {self.price} рублей'

ch = Chocolate(5)
ch.show_my_chocolate()
print(f'{ch}\n')

ch = Chocolate(6,'Апельсин')
ch.show_my_chocolate()
print(f'{ch}\n')

ch = Chocolate(3.5)
print(f'{ch}\n')

try:
  ch = Chocolate(0)
  print(ch)
except ValueError as err:
  print(err)
except:
  print('Ошибка')
finally:
  print()

try:
  ch = Chocolate(-1)
  print(ch)
except ValueError as err:
  print(err)
except:
  print('Ошибка')
print()

try:
  ch = Chocolate('a')
  print(ch)
except ValueError as err:
  print(err)
except:
  print('Ошибка')
print()

try:
  ch = Chocolate(7,1)
  print(ch)
  ch.show_my_chocolate()
except ValueError as err:
  print(err)
except:
  print('Ошибка')
```
